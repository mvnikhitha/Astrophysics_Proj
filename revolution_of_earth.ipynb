```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

E = -2
l = 3
e = np.sqrt(1+2*E/l**2)

def r(theta):
    return l**2 / (1-e*np.cos(theta))

def dtheta_dt(theta, t):
    return l/r(theta)**2

t = np.linspace(0, 10*(l**2 / (1- e**2))**(3 / 2)*(2*np.pi), 1000)
theta = odeint(dtheta_dt, 0, t)
plt.plot(t, theta)
plt.show()
```

```python
from matplotlib.animation import FuncAnimation
```

```python
fig,ax = plt.subplots()
ax.plot(r(theta)*np.cos(theta),r(theta)*np.sin(theta))
plt.show()
```

```python
planet, = ax.plot(0,0,marker = "o")
def  update(t):
  planet.set_data(r(theta[t])*np.cos(theta[t]),r(theta[t])*np.sin(theta[t]))
  return planet,

animation = FuncAnimation(fig,update,interval=10,frames=range(1000))
```

```python
animation.save('animation.mp4',writer='ffmpeg')
```
